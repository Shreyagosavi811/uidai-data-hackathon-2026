```python
import pandas as pd
import matplotlib.pyplot as plt

OUTPUT_PATH = "output"


def generate_visuals():
    raw = pd.read_csv(f"{OUTPUT_PATH}/raw_combined.csv")
    clean = pd.read_csv(f"{OUTPUT_PATH}/phase2_clean_dataset.csv")

    before = raw.shape[0]
    after = clean.shape[0]

    plt.figure(figsize=(6,4))
    plt.bar(['Before Cleaning', 'After Cleaning'], [before, after])
    plt.title('Aadhaar Records: Before vs After Cleaning')
    plt.ylabel('Number of Records')
    plt.tight_layout()
    plt.savefig(f"{OUTPUT_PATH}/before_after_cleaning.png")
    plt.show()

    print("ðŸ“Š Visualization generated")
```
