```python
import os
import pandas as pd

DATA_PATH = "data/api_data_aadhar_enrolment"
OUTPUT_PATH = "output"
CHUNK_SIZE = 200_000


def load_raw_data():
    print("üîç Searching for CSV files...")

    csv_files = []
    for root, _, files in os.walk(DATA_PATH):
        for f in files:
            if f.endswith(".csv"):
                csv_files.append(os.path.join(root, f))

    if not csv_files:
        raise FileNotFoundError("No CSV files found")

    print(f"üìÇ {len(csv_files)} CSV files detected")

    chunks = []
    for path in csv_files:
        print(f"üìÑ Loading {os.path.basename(path)}")
        for chunk in pd.read_csv(path, chunksize=CHUNK_SIZE):
            chunks.append(chunk)

    df = pd.concat(chunks, ignore_index=True)
    print("‚úÖ Combined shape:", df.shape)

    os.makedirs(OUTPUT_PATH, exist_ok=True)
    df.to_csv(f"{OUTPUT_PATH}/raw_combined.csv", index=False)

    return df
```
