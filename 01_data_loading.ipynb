Imports & Global **Setup**

```python
import pandas as pd
import numpy as np
from difflib import get_close_matches
```

Load Reference Dataset

```python
ref = pd.read_csv("reference_master.csv")
ref.columns = ref.columns.str.strip()

print(ref.columns.tolist())
```

Extract Required Reference Columns

```python
ref_subset = ref[['stateNameEnglish', 'localBodyNameEnglish', 'pincode']].copy()
ref_subset.columns = ['state', 'district', 'pincode']

ref_subset['state'] = ref_subset['state'].str.title().str.strip()
ref_subset['district'] = ref_subset['district'].str.title().str.strip()
ref_subset['pincode'] = ref_subset['pincode'].astype(str)

ref_subset.head()
```

Load Main Dataset

```python
df = pd.read_csv("main_data.csv")

df.columns = df.columns.str.lower().str.strip()

df['state'] = df['state'].str.title().str.strip()
df['district'] = df['district'].str.title().str.strip()
df['pincode'] = df['pincode'].astype(str)

df.head()
```

Load Age Dataset

```python
age_df = pd.read_csv("age_data.csv")

age_df.columns = age_df.columns.str.lower().str.strip()

age_df['state'] = age_df['state'].str.title().str.strip()
age_df['district'] = age_df['district'].str.title().str.strip()
age_df['pincode'] = age_df['pincode'].astype(str)

age_df.head()
```
