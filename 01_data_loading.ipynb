Imports & Global **Setup**

```python
import pandas as pd
import numpy as np
import os
from google.colab import drive

drive.mount('/content/drive')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
```

Load Reference Dataset

```python
DATA_PATH = "/content/drive/MyDrive/data_Aadhar/api_data_aadhar_enrolment"
OUTPUT_PATH = "/content/drive/MyDrive/data_Aadhar/output"

os.makedirs(OUTPUT_PATH, exist_ok=True)
```

Extract Required Reference Columns

```python
csv_files = sorted([f for f in os.listdir(DATA_PATH) if f.endswith(".csv")])

chunks = []
for file in csv_files:
    path = os.path.join(DATA_PATH, file)
    for chunk in pd.read_csv(path, chunksize=200_000):
        chunks.append(chunk)

df = pd.concat(chunks, ignore_index=True)

print("Combined shape:", df.shape)
df.head()
```

Load Reference(Geo) Dataset

```python
REF_PATH = os.path.join(DATA_PATH, "States_UTs.csv")

ref = pd.read_csv(REF_PATH)
ref.columns = ref.columns.str.strip()

ref = ref[['stateNameEnglish', 'localBodyNameEnglish', 'pincode']]
ref.columns = ['state', 'district', 'pincode']

ref['state'] = ref['state'].str.title().str.strip()
ref['district'] = ref['district'].str.title().str.strip()
ref['pincode'] = ref['pincode'].astype(str)

ref.drop_duplicates(inplace=True)
ref.head()
```

Load Age Dataset

```python
age_df = pd.read_csv("age_data.csv")

age_df.columns = age_df.columns.str.lower().str.strip()

age_df['state'] = age_df['state'].str.title().str.strip()
age_df['district'] = age_df['district'].str.title().str.strip()
age_df['pincode'] = age_df['pincode'].astype(str)

age_df.head()
```
